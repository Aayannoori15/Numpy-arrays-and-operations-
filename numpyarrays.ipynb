```python
#3d arrays 
```

```python
import numpy as np

arr3d = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])

print(arr3d)
```

```python
print(arr3d.sum(axis=2))
print(arr3d.prod(axis=0))
print(arr3d.min(axis=2))
print(arr3d.max(axis=1))
print(arr3d.argmin())
print(arr3d.argmax())
```

```python
#slicing 3d array 
```

```python
sliced3d=arr3d[1:2,1:2,1:2]
print(sliced3d)
```
